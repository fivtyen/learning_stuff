### Linear regression, torch

```python
%matplotlib inline
from fastai import *
import torch
import matplotlib.pyplot as plt
```

```python
n = 100
x = torch.ones(n,2)
x[:,0].uniform_(-1,1)
```

```python
a = torch.tensor([3.,2]);
y = x@a + torch.rand(n)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x[:,0], y);
```

```python
a = torch.tensor([1.,1.])
```

```python
def mse(y_hat, y):
    return((y_hat-y)**2).mean()
```

```python
a = torch.nn.Parameter(a)
```

```python
def update():
    y_hat = x@a
    loss = mse(y, y_hat)
    if t % 10 == 0: print(loss)
    loss.backward()
    with torch.no_grad():
        a.sub_(lr * a.grad)
        a.grad.zero_()
```

```python
lr = 1e-1
for t in range(100):
    update()
```

```python
a.requires_grad = False
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x[:,0], y);
plt.scatter(x[:,0], x@a);
```
